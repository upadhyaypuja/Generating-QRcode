```python
#importing libraries
import qrcode
import matplotlib.pyplot as plt
#generating image feature
code_f = qrcode.QRCode(box_size = 35, border = 4)
#adding data
code_f.add_data("https://themarketingboy.com/")
code_f.make(fit=True)
qrcode_img = code_f.make_image(fill_color = "black",back_color = "red")
#saving image
qrcode_img.save("qrcodeimg.png")
plt.imshow(qrcode_img,cmap='gray')
```
